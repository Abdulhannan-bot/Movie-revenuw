# Predict Movie revenue with Linear Regression by analysing budget against cost

```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

## Read Data 

```python
data = pandas.read_csv('cost_revenue_clean.csv')
print(data)
```

## Plot Worldwide Gross against Production Budget

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X,y,alpha=0.3)
plt.title('Film cost v/s Global revenue')
plt.xlabel('Production Cost $')
plt.ylabel('Worlwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,500000000)
plt.show()
```

## Linear Regression

```python
regression = LinearRegression()
regression.fit(X,y)
```

## Calculate the R-squared metric

```python
regression.coef_
regression.intercept_
regression.score(X,y)
```

## Plot the data with Linear Regression

```python
plt.figure(figsize=(12,8))
plt.scatter(X,y,alpha=0.3)
plt.plot(X.values,regression.predict(X), color = 'red', linewidth = 4)
plt.title('Film cost v/s Global revenue')
plt.xlabel('Production Cost $')
plt.ylabel('Worlwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,500000000)
plt.show()
```
